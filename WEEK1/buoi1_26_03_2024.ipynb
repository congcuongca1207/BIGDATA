### Đọc dữ liệu:
  1. spark.read.csv(...)
  2. spark.read.option('header', 'true').csv(...)

### Xem dữ liệu
  1. Xem qua: df.show()
  2. Xem kỹ
    - Dạng dữ liệu: .printSchema()
    - Tổng quát thống kê: .describe()

### Xử lý cột
  1. Thêm cột: .withColumn()
  2. Vứt cột: .drop()
  3. Đổi tên cột: .withColumnRenamed()
  4. Chọn cột: .select(['cot1', 'cot2',...])

### Lọc dữ liệu: .filter()
  1. <, > , <=, >=, ==
  2. hoặc: |
  3. và: &

### Tổng hợp dữ liệu
  1. .mean()
  2. .sum()
  3. .count()
  4. .max()
  5. .min()
  6. .groupBy('nganh').mean()

```python
df_pyspark = spark.read.csv("test1.csv")
df_pyspark.show()
# mat header
```

```python
df_pyspark = spark.read.option("header", "true").csv("test1.csv")
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
df_pyspark.describe().show()
```

# xử lý cột

```python
df_pyspark = df_pyspark.withColumn("Experience after 2 year", df_pyspark["Experience"] +2)
df_pyspark.show()
```

# xóa cột

```python
df_pyspark = df_pyspark.drop("Experience after 2 year")
df_pyspark.show()
```

# đổi tên cột

```python
df_pyspark = df_pyspark.withColumnRenamed('age', "Age")
df_pyspark.show()
```

# Chọn cột

```python
df_pyspark['Name', 'Age'].show()
```

# lọc dữ liệu

```python
df_pyspark.filter(df_pyspark['Salary'] <= 20000).show()
```

```python
# lọc nhỏ hơn 20000 hoặc lớn hơn 15000
df_pyspark.filter((df_pyspark['Salary'] <= 20000) |
                  (df_pyspark['Salary'] >= 15000)).show()
```
