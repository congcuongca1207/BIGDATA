## 5. Tên của người được trả cao nhất (bao gồm benefits) là gì?


### em đang dùng pandas để lấy ra giá trị lớn nhất

```python
import pandas as pd
pdf = df.toPandas()
total_max = pdf['TotalPayBenefits'].max()
total_max
```

```python
df.filter(df.TotalPayBenefits == total_max).select('EmployeeName').show()
```

## 6. Tên vị trí công việc (Job Title) của người đó là gì?


```python
df.filter(df.TotalPayBenefits == total_max).select('JobTitle').show()
```

## 7. Tên của người được trả thấp nhất (bao gồm benefits) là gì?

```python
from pyspark.sql.functions import min
minPaid = df.agg(min('TotalPayBenefits')).collect()[0][0]
df.filter(df['TotalPayBenefits'] == minPaid).select('EmployeeName').show()
```
